# B. Manipulation d’un jeu de données

```python
# 1
iris = datasets.load_iris()
```

```python
# 2
donnees = iris.data.copy()
print(donnees)

nom_variables = iris.feature_names.copy()
print(nom_variables)

nom_classes = iris.target_names.copy()
print(nom_classes)
```

```python
# 3
from prettytable import PrettyTable

table = PrettyTable(["Données","Nom de Classe"])
classes = iris.target

for i in range(donnees.shape[0]):
    table.add_row([str(donnees[i]),str(nom_classes[classes[i]])])   
print(table)
```

```python
# 4
means = donnees.mean(0)
stds = donnees.std(0)
mins = donnees.min(0)
maxs = donnees.max(0)

rows = [""]
rows.extend(nom_variables)

table = PrettyTable(rows)
table.add_row(["means",means[0],means[1],means[2],means[3]])
table.add_row(["stds",stds[0],stds[1],stds[2],stds[3]])
table.add_row(["mins",mins[0],mins[1],mins[2],mins[3]])
table.add_row(["maxs",maxs[0],maxs[1],maxs[2],maxs[3]])
print(table)
```

```python
# 5
table = PrettyTable(["","Nombre de Données","Nombre de Variables"])
table.add_row(["size",donnees.size/donnees[0].size,donnees[0].size])
table.add_row(["shape",donnees.shape[0],donnees[0].shape[0]])
print(table)

print("le nombre de classes est : ",np.unique(classes).shape[0])
```

```python
# 2
mnist_data = mnist.data
mnist_target = mnist.target

print("Données: \n",mnist_data)

table = PrettyTable(["","nombre_données","nombre_variables","nombre_classes"])
table.add_row(["size",mnist_data.size/mnist_data[0].size,mnist_data[0].size,mnist_target.size])
print(table)

print("Moyenne de chaque variable:\n",mnist_data.mean(0))
print("L'écart type de chaque variable:\n",mnist_data.std(0))
print("Min de chaque variable:\n",mnist_data.min(0))
print("Max de chaque variable:\n",mnist_data.max(0))

print("Nombre de classes (distinct) est: ",np.unique(mnist_target).size)
```

```python
# 2
from sklearn.datasets.samples_generator import make_blobs
X, Y = make_blobs(n_samples=1000, centers=4, n_features=2,random_state=1)
print("Nombre de données:",X.shape[0])
print("Les classes :",np.unique(Y))
```

```python
# 3
plt.figure()
plt.scatter(X[:,0],X[:,1],c=Y)
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("figure")
```

```python
# 4
def draw(X,Y,title):
    plt.figure()
    plt.scatter(X[:,0],X[:,1],c=Y)
    plt.xlim(-15,15)
    plt.ylim(-15,15)
    plt.xlabel("xlabel")
    plt.ylabel("ylabel")
    plt.title(title)
    plt.show()
    
Z,U = make_blobs(n_samples=100,n_features=2,centers=2,random_state=1)
V,W = make_blobs(n_samples=500,n_features=2,centers=3,random_state=1)
ZV = np.vstack((Z,V))
UW = np.hstack((U,W))
draw(Z,U,"100 données réparties en 2 groupes")
draw(V,W,"500 données réparties en 3 groupes")
draw(ZV,UW,"Concaténation de 2 jeux données ")
```
